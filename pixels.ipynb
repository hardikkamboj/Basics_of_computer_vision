```python
import numpy as np
```

```python
img = np.random.random(size = (20,20,3))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
def separate_channels(img):
  shape = img.shape

  red = np.zeros(shape)
  red[:,:,0] = img[:,:,0]

  green = np.zeros(shape)
  green[:,:,1] = img[:,:,1]

  blue = np.zeros(shape)
  blue[:,:,2] = img[:,:,2]
  
  plt.figure(figsize = (30,8))

  plt.subplot(1,3,1)
  plt.imshow(red)

  plt.subplot(1,3,2)
  plt.imshow(green)

  plt.subplot(1,3,3)
  plt.imshow(blue)
```

```python
separate_channels(img)
```

```python
plt.figure(figsize = (8,6))
plt.imshow(img)
```

```python
# changing some values
```

```python
img[:,1,0] = 255
img[:,1,1] = 0
img[:,1,2] = 0
```

```python
plt.figure(figsize = (8,6))
plt.imshow(img)
```

```python
separate_channels(img)
```

```python
img[:,:,1] = 200
img[:,:,1] = np.clip(img[:,:,1],0,255)
```

```python
plt.imshow(img)
```

```python
separate_channels(img)
```
